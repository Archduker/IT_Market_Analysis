```python
import os
import re
import ast
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.options.display.float_format = '{:,.0f}'.format

#  1) Load file 
df = pd.read_csv("Jobs_IT.csv", low_memory=False)
print("Số dòng:", len(df))

#  2) Xác định cột 
def find_col(df, possible_names):
    low_map = {c.lower(): c for c in df.columns}
    for name in possible_names:
        if name in df.columns: return name
        if name.lower() in low_map: return low_map[name.lower()]
    for c in df.columns:
        if any(name.lower() in c.lower() for name in possible_names):
            return c
    return None

col_min     = find_col(df, ["salary_min","min_salary","luong_min","lương_min"])
col_max     = find_col(df, ["salary_max","max_salary","luong_max","lương_max"])
col_industry= find_col(df, ["industry","ngành","nganh","category","chuyên môn","chuyen mon"])
col_city    = find_col(df, ["city","thành phố","thanh pho","location","tỉnh","tinh","province"])
col_level   = find_col(df, ["level","cấp bậc","seniority","position level"])
col_skills  = find_col(df, ["skills","kỹ năng","skill","ky nang"])
col_source  = find_col(df, ["source","nguồn","platform","jobsite","nguon"])

#  3) Chuẩn hoá lương 
df["_salary_min"] = pd.to_numeric(df[col_min], errors="coerce")
df["_salary_max"] = pd.to_numeric(df[col_max], errors="coerce")
df.loc[df["_salary_min"]<=0, "_salary_min"] = np.nan
df.loc[df["_salary_max"]<=0, "_salary_max"] = np.nan
df.loc[df["_salary_min"].notna() & df["_salary_max"].isna(), "_salary_max"] = df["_salary_min"]
df.loc[df["_salary_max"].notna() & df["_salary_min"].isna(), "_salary_min"] = df["_salary_max"]

# Đảm bảo min <= max
mask_swap = df["_salary_min"] > df["_salary_max"]
df.loc[mask_swap, ["_salary_min","_salary_max"]] = df.loc[mask_swap, ["_salary_max","_salary_min"]].values

df["_salary_mid"] = (df["_salary_min"] + df["_salary_max"]) / 2
df["_has_salary"] = df["_salary_min"].notna() & df["_salary_max"].notna()

#  4) Tổng quan 
overall = {
    "rows_total": len(df),
    "rows_with_salary": int(df["_has_salary"].sum()),
    "share_with_salary_%": round(100*df["_has_salary"].mean(), 2),
    "min_salary(VND)": df.loc[df["_has_salary"], "_salary_min"].min(),
    "avg_mid_salary(VND)": df.loc[df["_has_salary"], "_salary_mid"].mean(),
    "median_mid_salary(VND)": df.loc[df["_has_salary"], "_salary_mid"].median(),
    "max_salary(VND)": df.loc[df["_has_salary"], "_salary_max"].max(),
}
overall_df = pd.DataFrame([overall])
print("== Tổng quan lương ==")
display(overall_df)

# 5) Hàm thống kê theo nhóm  
df["__dummy__"] = 1

def group_stats(group_col, top_n=15):
    if group_col is None: 
        return pd.DataFrame()
    g = (df.groupby(group_col, dropna=False)
           .agg(
               postings=("__dummy__","sum"),
               disclosed=("_has_salary","sum"),
               disclosed_share=("_has_salary","mean"),
               mean_mid_vnd=("_salary_mid","mean"),
               median_mid_vnd=("_salary_mid","median"),
               min_min_vnd=("_salary_min","min"),
               max_max_vnd=("_salary_max","max")
           ).reset_index())
    g["disclosed_share"] = (g["disclosed_share"]*100).round(2)
    g = g.sort_values(["disclosed","mean_mid_vnd"], ascending=[False, False]).head(top_n)
    ordered = [group_col,"postings","disclosed","disclosed_share",
               "mean_mid_vnd","median_mid_vnd","min_min_vnd","max_max_vnd"]
    return g[ordered]

by_industry = group_stats(col_industry)
by_city     = group_stats(col_city)
by_level    = group_stats(col_level)
by_source   = group_stats(col_source)

print("\n== Top ngành công khai lương ==")
display(by_industry)
print("\n== Top thành phố ==")
display(by_city)
print("\n== Top cấp bậc ==")
display(by_level)
print("\n== Top nguồn đăng ==")
display(by_source)

#  6) Theo kỹ năng 
def explode_skills(df, col_skills):
    if col_skills is None: return pd.DataFrame()
    tmp = df[[col_skills,"_salary_mid","_has_salary","__dummy__"]].copy()
    tmp[col_skills] = (tmp[col_skills].astype(str)
                       .str.replace(r"[|;]|／|、", ",", regex=True))
    tmp = tmp.assign(skill=tmp[col_skills].str.split(",")).explode("skill")
    tmp["skill"] = tmp["skill"].astype(str).str.strip().str.lower()
    tmp = tmp[(tmp["skill"].notna()) & (tmp["skill"]!="") & (tmp["skill"]!="nan")]
    g = (tmp.groupby("skill", dropna=False)
           .agg(postings=("__dummy__","sum"),
                disclosed=("_has_salary","sum"),
                mean_mid_vnd=("_salary_mid","mean"))
           .reset_index()
           .sort_values(["disclosed","mean_mid_vnd"], ascending=[False, False])
           .head(30))
    return g

by_skill = explode_skills(df, col_skills)
print("\n== Top kỹ năng ==")
display(by_skill)

#  7) Biểu đồ 
plt.rcParams["figure.figsize"] = (10,5)

if not by_industry.empty:
    plt.barh(by_industry[col_industry].astype(str), by_industry["disclosed"])
    plt.title("Top ngành công khai lương nhiều nhất")
    plt.xlabel("Số tin có công khai lương")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()

if not by_source.empty:
    plt.barh(by_source[col_source].astype(str), by_source["disclosed"])
    plt.title("Top nguồn công khai lương nhiều nhất")
    plt.xlabel("Số tin có công khai lương")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```
